### f. Univariate and Multivariate

```python
# Set style for plots
sns.set_style("whitegrid")
```

- Display the distribution plot of the price

```python
# Display the distribution of the price
sns.histplot(df.price)
plt.title("Distribution of Price")
plt.xlabel("Price")
plt.show()
```

```python
# Create a new column log_price
df.loc[:,["log-price"]] = np.log1p(df.price)
df
```

```python
# Display the distribution plot of the log_price
sns.histplot(df['log-price'])
plt.title("Distribution of Log price")
plt.xlabel("Log_price")
plt.show()
```

```python
# Drop the original column
df = df.drop(columns="price")
df
```

- Find the records of where the log_price is smaller than its 1% percentile and greater than 99% percentile

```python
lower_bound = df['log-price'].quantile(0.01)
upper_bound = df['log-price'].quantile(0.99)
df = df[(df['log-price'] >= lower_bound) & (df['log-price'] <= upper_bound)]
df
```

```python
# Display the distribution plot of the log_price again
sns.histplot(df['log-price'])
plt.title("Distribution of Log Price After Removing Outliers")
plt.xlabel("Log_price")
plt.show()
```

### g. Multivariate analysis

- Display a multivariate analysis plot for log_price against mileage

```python
sns.scatterplot(x=df['sqrt-mileage'], y=df['log-price'])
plt.title('Multivariate Analysis: Log Price vs Square Root Mileage')
plt.xlabel('Square Root Mileage')
plt.ylabel('Log Price')
plt.show()
```

Observation:
The data points are widely distributed, and there is a weak negative correlation between the `log_price` and `sqrt_mileage`.

1. **Dense Distribution**: The majority of the data points are clustered in the upper range of `log_price` (between 9 and 11), suggesting that most cars in the dataset are priced within a specific range.
2. **Weak Negative Trend**: There is a slight tendency for higher `sqrt_mileage` values to be associated with lower `log_price`, indicating that cars with more mileage tend to have lower prices.
3. **Outliers**: Some points at both the lower and upper ends of `sqrt_mileage` exhibit deviations from the main trend, suggesting that some high-mileage cars still maintain high prices.

- Display a multivariate analysis plot for log_price against year

```python
sns.scatterplot(x=df['year'], y=df['log-price'])
plt.title('Multivariate Analysis: Log Price vs Year')
plt.xlabel('Year')
plt.ylabel('Log Price')
plt.show()
```

Observation: Log Price and Year

There is a noticeable trend where newer cars tend to have higher `log_price`. Older cars, particularly those from 2010-2015, generally exhibit lower log-transformed prices compared to those from 2016 onwards. This makes sense, as newer cars are generally more expensive due to better technology, lower wear and tear, and higher demand.


### h. Reducing the number of values in categorical columns

- Check the value counts for each categorical column

```python
print(df.transmission.value_counts())
print()
print(df['fuel-type'].value_counts())
print()
print(df.make.value_counts())
print()
print(df.model.value_counts())
```

```python
# Drop the model column
df = df.drop(columns="model")
df
```

```python
make_mean = df.make.value_counts().mean()
filter = df.make.value_counts() < make_mean
make_other = filter[filter == True].index.to_list()

#Replace with other
df.make= df.make.apply(lambda x: 'other' if x in make_other else x)

#Check the updated value counts
df.make.value_counts()
```
