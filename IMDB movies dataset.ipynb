```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Load data
df = pd.read_csv('imdb_top_1000.csv')
df.head()
```

```python
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Handling missing values
df.Certificate = df.Certificate.fillna('Unrated')
df['Meta_score']= df['Meta_score'].fillna(df['Meta_score'].mode()[0])
df.Gross = df.Gross.fillna(df.Gross.mode()[0])

# Check null values again
df.isnull().sum().sum()
```

```python
# Clean and convert 'Gross' to numeric
df['Gross'] = df['Gross'].astype(str).str.replace(',', '', regex=False)
df['Gross'] = pd.to_numeric(df['Gross'], errors='coerce')

# Clean and convert 'Runtime' to numeric
df['Runtime'] = df['Runtime'].astype(str).str.replace(' min', '', regex=False)
df['Runtime'] = pd.to_numeric(df['Runtime'], errors='coerce')

# Convert 'Released_Year' to numeric
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce')
```

```python
# Split genres
df['Genre'] = df['Genre'].astype(str)
genre_content = df.assign(genre=df.Genre.str.split(',')).explode('Genre')
genre_content['Genre'] =  genre_content['Genre'].str.strip()
genre_content.head()
```
